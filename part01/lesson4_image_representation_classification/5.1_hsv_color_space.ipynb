```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
# reading and visualize image
img = cv2.imread('images/water_balloons.jpg')
print (img.shape)

plt.imshow(img)
```

```python
# convert image from opencv BGR to RGB format 
img_rgb = img.copy()

dst_rgb = cv2.cvtColor(img_rgb, cv2.COLOR_BGR2RGB)
plt.imshow(dst_rgb)
```

```python
#split RGB image to R,G,B channels
(R, G, B) = cv2.split(dst_rgb)
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 15))
ax1.set_title("red")
ax1.imshow(R, cmap = 'gray')
ax2.set_title("green")
ax2.imshow(G, cmap = 'gray')
ax3.set_title("blue")
ax3.imshow(B, cmap = 'gray')
```

```python
#convert RGB to HSV color space
dst_rgb_copy = dst_rgb.copy()

hsv_img = cv2.cvtColor(dst_rgb_copy, cv2.COLOR_RGB2HSV)
plt.imshow(hsv_img)
```

```python
#split HSV image to H,S,V channels
(H, S, V) = cv2.split(hsv_img)
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 15))
ax1.set_title("hue")
ax1.imshow(H, cmap = 'gray')
ax2.set_title("saturation")
ax2.imshow(S, cmap = 'gray')
ax3.set_title("value")
ax3.imshow(V, cmap = 'gray')
```

```python
# Define our color selection criteria in RGB values
lower_pink = np.array([180,0,100]) 
upper_pink = np.array([255,255,230])
```

```python
# Define the masked area in RGB space
mask_rgb = cv2.inRange(dst_rgb_copy, lower_pink, upper_pink)

# mask the image
masked_image = np.copy(dst_rgb_copy)
masked_image[mask_rgb==0] = [0,0,0]

# Vizualize the mask
plt.imshow(masked_image)
```

```python
# Define our color selection criteria in HSV values
lower_hue = np.array([150,0,0]) 
upper_hue = np.array([180,255,255])
```

```python
# Define the masked area in HSV space
mask_hsv = cv2.inRange(hsv_img, lower_hue, upper_hue)

# mask the image
masked_image = np.copy(hsv_img)
masked_image[mask_hsv==0] = [0,0,0]

# Vizualize the mask
plt.imshow(masked_image)
```
